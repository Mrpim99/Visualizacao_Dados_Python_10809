## üóÇÔ∏è Contextualiza√ß√£o do Dataset: Iris

O dataset **Iris** √© um dos conjuntos de dados mais conhecidos em estat√≠stica e ci√™ncia de dados. Foi publicado pelo estat√≠stico brit√¢nico **Ronald A. Fisher** em 1936, e cont√©m informa√ß√µes sobre 150 amostras de flores da esp√©cie *Iris*, divididas em tr√™s subesp√©cies:

- *Iris setosa*
- *Iris versicolor*
- *Iris virginica*

Para cada flor, foram recolhidas quatro medidas:

- **Comprimento da s√©pala** (*sepal_length*)
- **Largura da s√©pala** (*sepal_width*)
- **Comprimento da p√©tala** (*petal_length*)
- **Largura da p√©tala** (*petal_width*)

Estas medidas est√£o expressas em cent√≠metros (cm) e permitem distinguir facilmente entre as esp√©cies com recurso a visualiza√ß√µes.

Este dataset √© ideal para aprender a utilizar bibliotecas de visualiza√ß√£o como o **Matplotlib**, explorando rela√ß√µes entre vari√°veis, distribui√ß√µes e compara√ß√µes entre grupos.


```python
# üì¶ Importa√ß√£o de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# üì• Carregamento do dataset
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# üëÄ Visualiza√ß√£o inicial dos dados
df.head()
```

## üìä Exemplo 1: Gr√°fico de Dispers√£o

Visualiza√ß√£o da rela√ß√£o entre duas vari√°veis num√©ricas com um gr√°fico de dispers√£o (scatter plot).


```python
# Rela√ß√£o entre comprimento e largura da p√©tala
plt.figure(figsize=(6, 4))
plt.scatter(df['petal_length'], df['petal_width'], color='blue')
plt.title("Comprimento vs Largura da P√©tala")
plt.xlabel("Comprimento da P√©tala (cm)")
plt.ylabel("Largura da P√©tala (cm)")
plt.grid(True)
plt.show()
```

### üìù Exerc√≠cio 1:
Cria um gr√°fico de dispers√£o para analisar a rela√ß√£o entre o **comprimento e a largura da s√©pala**.
Inclui os r√≥tulos dos eixos e o t√≠tulo do gr√°fico.


```python
plt.figure(figsize=(6, 4))
plt.scatter(df['sepal_length'], df['sepal_width'], color='blue')
plt.title("Comprimento vs Largura da P√©tala")
plt.xlabel("Comprimento da P√©tala (cm)")
plt.ylabel("Largura da P√©tala (cm)")
plt.grid(True)
plt.show()
```

```python
 #Cores por esp√©cie
cores = {
    'setosa': 'blue',
    'versicolor': 'green',
    'virginica': 'red'
}

plt.figure(figsize=(6, 4))

# Plotar cada esp√©cie separadamente
for especie in df['species'].unique():
    subset = df[df['species'] == especie]
    plt.scatter(subset['sepal_length'], subset['sepal_width'],
                color=cores[especie], label=especie, edgecolors='black', s=40)

plt.title("Comprimento vs Largura da S√©pala por Esp√©cie")
plt.xlabel("Comprimento da S√©pala (cm)")
plt.ylabel("Largura da S√©pala (cm)")
plt.grid(True)
plt.legend(title="Esp√©cie")
plt.show()
```

#### ‚úÖ Resolu√ß√£o do Exerc√≠cio 1:
Vamos criar um gr√°fico de dispers√£o entre o comprimento e a largura da s√©pala.
1. Utilizamos `plt.scatter()` para gerar o gr√°fico.
2. Usamos os campos `sepal_length` e `sepal_width` como vari√°veis.
3. Adicionamos r√≥tulos e t√≠tulo para legibilidade.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['sepal_length'], df['sepal_width'], color='teal')
plt.title("Comprimento vs Largura da S√©pala")
plt.xlabel("Comprimento da S√©pala (cm)")
plt.ylabel("Largura da S√©pala (cm)")
plt.grid(True)
plt.show()
```

## üìà Exemplo 2: Gr√°fico de Linhas

Visualiza√ß√£o de valores agregados (como m√©dias) por categoria usando linhas.


```python
# M√©dia do comprimento da p√©tala por esp√©cie
medias = df.groupby("species")["petal_length"].mean()

plt.figure(figsize=(6, 4))
plt.plot(medias.index, medias.values, marker='o', linestyle='-', color='green')
plt.title("M√©dia do Comprimento da P√©tala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento M√©dio da P√©tala (cm)")
plt.grid(True)
plt.show()
```

### üìù Exerc√≠cio 2:
Cria um gr√°fico de linhas que represente a **m√©dia da largura da s√©pala** por esp√©cie.
Inclui marcadores nos pontos e personaliza o gr√°fico.


#### ‚úÖ Resolu√ß√£o do Exerc√≠cio 2:
1. Agrupamos o dataset pela coluna `species`.
2. Calculamos a m√©dia de `sepal_width`.
3. Usamos `plt.plot()` com marcadores para representar as m√©dias.

```python
media_sepala = df.groupby("species")["sepal_width"].mean()
plt.figure(figsize=(6, 4))
plt.plot(media_sepala.index, media_sepala.values, marker='s', linestyle='--', color='darkorange')
plt.title("M√©dia da Largura da S√©pala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Largura M√©dia da S√©pala (cm)")
plt.grid(True)
plt.show()
```

üîç O que foi adicionado:
plt.text(...) escreve acima de cada ponto:

Nome da esp√©cie (especie)

Valor m√©dio com 2 casas decimais ({valor:.2f})

O \n quebra a linha entre nome e valor

tight_layout() ajuda a evitar cortes nos r√≥tulos



```python
media_sepala = df.groupby("species")["sepal_width"].mean()

plt.figure(figsize=(6, 4))
plt.plot(media_sepala.index, media_sepala.values, marker='s', linestyle='--', color='darkorange')

# Adicionar texto sobre cada ponto
for i, especie in enumerate(media_sepala.index):
    valor = media_sepala.values[i]
    plt.text(i, valor + 0.05, f"{especie}\n{valor:.2f}", ha='center', va='bottom')

plt.title("M√©dia da Largura da S√©pala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Largura M√©dia da S√©pala (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### üìù Exerc√≠cio 3:
Cria um gr√°fico de barras com a **m√©dia do comprimento da s√©pala** por esp√©cie.
Adiciona t√≠tulo e r√≥tulos nos eixos.


#### ‚úÖ Resolu√ß√£o do Exerc√≠cio 3:
1. Agrupamos os dados por esp√©cie.
2. Calculamos a m√©dia de `sepal_length`.
3. Representamos com `plt.bar()`.

```python
media_sepal_length = df.groupby("species")["sepal_length"].mean()
plt.figure(figsize=(6, 4))
plt.bar(media_sepal_length.index, media_sepal_length.values, color='skyblue')
plt.title("Comprimento M√©dio da S√©pala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento M√©dio da S√©pala (cm)")
plt.show()
```

üß† O que acontece aqui:
bar.get_height() devolve o valor da barra

plt.text(...) posiciona o texto centrado no topo

+ 1 d√° um pequeno espa√ßo acima da barra para o texto n√£o ficar colado

```python
media_sepal_length = df.groupby("species")["sepal_length"].mean()

plt.figure(figsize=(6, 4))
bars = plt.bar(media_sepal_length.index, media_sepal_length.values, color='skyblue')

# Adicionar valores em cima de cada barra
for bar in bars:
    altura = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, altura + 0.05,
             f"{altura:.2f}", ha='center', va='bottom', fontsize=10)

plt.title("Comprimento M√©dio da S√©pala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento M√©dio da S√©pala (cm)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Aqui est√° a vers√£o com os valores mostrados dentro das barras (em vez de por cima), e ainda com cores diferentes para cada esp√©cie no texto, usando apenas matplotlib

Tamb√©m:
üîç Notas:
f"{altura:.2f}": mostra o valor com 2 casas decimais

+ 0.05: espa√ßo acima da barra

ha='center': centraliza o texto

```python
media_sepal_length = df.groupby("species")["sepal_length"].mean()

# Cores por esp√©cie (usadas tanto na barra como no texto)
cores = {
    'setosa': 'steelblue',
    'versicolor': 'mediumseagreen',
    'virginica': 'tomato'
}

plt.figure(figsize=(6, 4))
bars = plt.bar(media_sepal_length.index, media_sepal_length.values,
               color=[cores[especie] for especie in media_sepal_length.index])

# Adicionar valores DENTRO das barras com a cor correspondente
for bar in bars:
    altura = bar.get_height()
    especie = bar.get_x() + bar.get_width()/2
    plt.text(especie, altura - 0.2,  # posiciona mais abaixo
             f"{altura:.2f}", 
             ha='center', va='top', 
             fontsize=10, color='white', fontweight='bold')

plt.title("Comprimento M√©dio da S√©pala por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento M√©dio da S√©pala (cm)")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

## üìä Exemplo 4: Histograma

Visualiza√ß√£o da distribui√ß√£o de uma vari√°vel cont√≠nua.


```python
# Distribui√ß√£o do comprimento da p√©tala
plt.figure(figsize=(6, 4))
plt.hist(df["petal_length"], bins=15, color='orange', edgecolor='black')
plt.title("Distribui√ß√£o do Comprimento da P√©tala")
plt.xlabel("Comprimento da P√©tala (cm)")
plt.ylabel("Frequ√™ncia")
plt.show()
```

### üìù Exerc√≠cio 4:
Cria um histograma para representar a **distribui√ß√£o da largura da s√©pala**.
Usa pelo menos 10 intervalos (bins).


#### ‚úÖ Resolu√ß√£o do Exerc√≠cio 4:
1. Selecionamos a coluna `sepal_width`.
2. Criamos um histograma com `plt.hist()`.
3. Definimos `bins=10` conforme solicitado.

```python
plt.figure(figsize=(6, 4))
plt.hist(df["sepal_width"], bins=10, color='salmon', edgecolor='black')
plt.title("Distribui√ß√£o da Largura da S√©pala")
plt.xlabel("Largura da S√©pala (cm)")
plt.ylabel("Frequ√™ncia")
plt.show()
```

üîç Explica√ß√£o:
plt.hist(...) retorna:

contagens: n¬∫ de ocorr√™ncias por bin

bins: limites de cada intervalo

patches: as barras desenhadas

plt.text(...): escreve o n√∫mero no centro de cada barra

```python
plt.figure(figsize=(6, 4))

# Criar histograma e capturar as barras
contagens, bins, patches = plt.hist(df["sepal_width"], bins=10, color='salmon', edgecolor='black')

# Adicionar valores em cima de cada barra
for i in range(len(patches)):
    altura = patches[i].get_height()
    if altura > 0:
        plt.text(patches[i].get_x() + patches[i].get_width()/2, altura + 0.5,
                 str(int(altura)), ha='center', va='bottom', fontsize=10)

plt.title("Distribui√ß√£o da Largura da S√©pala")
plt.xlabel("Largura da S√©pala (cm)")
plt.ylabel("Frequ√™ncia")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

## üß† Desafio Final

### üèÅ Desafio:
Cria um gr√°fico que combine m√∫ltiplos tipos (por exemplo, barras e linhas) para representar **duas vari√°veis diferentes por esp√©cie** (ex: comprimento e largura da p√©tala).  
Podes usar `plt.twinx()` ou criar duas figuras lado a lado.


#### ‚úÖ Resolu√ß√£o do Desafio Final:
1. Calculamos duas m√©dias por esp√©cie: `petal_length` e `petal_width`.
2. Representamos uma com barras e outra com linha.
3. Usamos `twinx()` para criar dois eixos y independentes no mesmo gr√°fico.

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

especies = df['species'].unique()
media_petal_length = df.groupby('species')['petal_length'].mean()
media_petal_width = df.groupby('species')['petal_width'].mean()

# Barras para comprimento
ax1.bar(especies, media_petal_length, color='lightblue', label='Comprimento')
ax1.set_xlabel('Esp√©cie')
ax1.set_ylabel('Comprimento M√©dio da P√©tala', color='blue')

# Linha para largura
ax2 = ax1.twinx()
ax2.plot(especies, media_petal_width, color='red', marker='o', label='Largura')
ax2.set_ylabel('Largura M√©dia da P√©tala', color='red')

plt.title("Compara√ß√£o: Comprimento vs Largura da P√©tala por Esp√©cie")
plt.grid(True)
plt.show()
```

‚úÖ Resultado:
A legenda mostra: Comprimento (barras) + Largura (linha)

A legenda aparece no canto superior esquerdo

Tudo continua 100% com matplotlib

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(8, 5))

# Obter m√©dias
media_petal_length = df.groupby('species')['petal_length'].mean()
media_petal_width = df.groupby('species')['petal_width'].mean()

# Cores por esp√©cie
cores = {
    'setosa': 'steelblue',
    'versicolor': 'mediumseagreen',
    'virginica': 'darkorange'
}

# Criar barras com cor e legenda
barras = []
labels = []
for especie in media_petal_length.index:
    bar = ax1.bar(especie, media_petal_length[especie],
                  color=cores[especie])
    barras.append(bar[0])  # bar √© um container
    labels.append(f'{especie} (comprimento)')

ax1.set_xlabel('Esp√©cie')
ax1.set_ylabel('Comprimento M√©dio da P√©tala', color='blue')

# Criar linha no eixo secund√°rio
ax2 = ax1.twinx()
linha, = ax2.plot(media_petal_width.index, media_petal_width.values,
                  color='red', marker='o', label='Largura m√©dia')
ax2.set_ylabel('Largura M√©dia da P√©tala', color='red')

# Adicionar legenda (barras + linha)
handles = barras + [linha]
labels += ['Largura m√©dia']
ax1.legend(handles, labels, loc='upper left')

plt.title("Comprimento vs Largura da P√©tala por Esp√©cie")
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
